```python
#non keyword variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
  for i in args:
    print(i)




```

```python
test(1,2,3,4,5,6)
```

```python
def sum_number(*args):
  sum=0
  for num in args:
    sum+=num




  return sum





```

```python
sum_number(1,2,3,4,5,6)
```

```python
#keyword variable argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a=10,b=20,c=30)
```

```python
import matplotlib.pyplot as plt
import numpy as np
cgpa=np.array([6,7,8,9,10])
package=np.array([4,5,6,7,9])
plt.plot(cgpa,package,"*:g",ms=13, mec = "red",mfc ="orange",linewidth=3)
plt.xlabel("cgpa")
plt.ylabel("package in lpa")
plt.title("Shoolini university students data")
plt.grid()
plt.show()
```

```python
#bar plot- it work on numrical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit=np.array(["Apple","Banana","Orange","watermelon"])
quantity=np.array([100,120,80,70])
color=np.array(["red","yellow","orange","green"])
plt.bar(fruit,quantity,color=color)
plt.xlabel("fruit")
plt.ylabel("quantity")
plt.title("fruits data")
plt.show()
```

```python
#bar plot- it work on numrical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit=np.array(["Apple","Banana","Orange","watermelon"])
quantity=np.array([100,120,80,70])
color=np.array(["red","yellow","orange","green"])
plt.barh(fruit,quantity,color=color)
plt.xlabel("quantity")
plt.ylabel("fruit")
plt.title("fruits data")
plt.show()
```

```python
#pie plot
percentage_of_fruit = np.array([20,25,30,25])
fruit = np.array(["Apple","Banana","Orange","watermelon"])
myexplode=np.array([0.2,0,0,0])
plt.pie(percentage_of_fruit,labels=fruit,explode=myexplode)
plt.legend(title="fruit",loc="best")
plt.show()
```
